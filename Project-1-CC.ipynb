### Challenge 1.1: Largest Palindrome
A palindromic number reads the same both ways. For example, 1234321 is a palindrome. The largest palindrome made from the product of two two-digit numbers is 9009 = 91 Ã— 99. Find the largest palindrome made from the product of two three-digit numbers.

```python
answerlist = []

for x in range(101,1000):
    for y in range(101,1000):
        answer = str(x*y)
        if len(str(answer))==5:
            if answer[0:1]==answer[-1] and answer[1:2]==answer[-2]:
                print str(x) +"x"+ str(y) +" "+answer
        if len(str(answer))==6:
            if answer[0:1]==answer[-1] and answer[1:2]==answer[-2] and answer[2:3]==answer[-3]:
                print str(x) +"x"+ str(y) +" "+answer
                answerlist.append(answer)

print "Answer:" +  str(max(answerlist))
```

### Challenge 1.3: Multiples of 3 and 5
If we list all of the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6, and 9. The sum of these multiples is 23. Find the sum of all the multiples of 3 and 5 below 1,000.

```python
multiples = set()
multipletotal = 0

for m3 in range(1,1000,3):
    multiples.add(m3)
for m5 in range(1,1000,5):
    multiples.add(m5)
    
for item in multiples:
    multipletotal = multipletotal + item

print multiples
print "Answer:" + str(multipletotal)
```
